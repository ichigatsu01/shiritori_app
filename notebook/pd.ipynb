```python
# kyoikukihongoi_2009Bを加工するためのファイル
import pandas as pd
```

```python
# 使おうとしたkyoikukihongoi_2009B.csvがshift-jisではなかったため、文字コードを調査する
# pip import chardetが必要
import chardet
with open('../data/kyoikukihongoi_2009B.csv', 'rb') as f:
    res = chardet.detect(f.read(5000))
    print(res)
```

```python
# engine=python...読込エンジンを厳格なC言語ベースではなく、Pythonベースに変更
# on_bad_lines='warn'...デフォルトでは'error'、異常な行があるとエラーをはいて中断してしまう。
# これを'warn'に変えることで警告を出しつつも処理をスキップするように変更
csvfile = pd.read_csv('../data/kyoikukihongoi_2009B.csv', encoding='CP932', engine='python', on_bad_lines='warn')
df = pd.DataFrame(data=csvfile)
```

```python
df
```

```python
print(df.loc[0])
```

```python
# 最終行はブランクのようなので除外する
df2 = df[['見出し', '品詞', '出現数']].loc[:27201]
df2
```

```python
# pip install jaconv
# 見出しの文字をすべてひらがなに変える
import jaconv
df2['見出し'] = df2['見出し'].apply(jaconv.kata2hira)
df2
```

```python
# 品詞が「名」になっている行だけを抽出
df3 = df2.loc[df2['品詞'] == '名']
df3
```

```python
# 出現数が4以上のものに限定する
df4 = df3.loc[df3['出現数'] >= 4]
df4
```

```python
# 「ん」で終わる単語を除外
# df['...'].str.endswith → 項目
df5 = df4.loc[~df4['見出し'].str.endswith('ん')]
df5
```

```python
# 番外：sliceとendswithの比較
# sliceの方がシンプルで少し早いけどやや固定的、空白文字はエラー
# endswithは空文字でもエラーを吐かない（falseを返す）、pandasでもstrを付ければ使える、ちょっとだけ重い
s = "ごはん"
print(s[-1] == "ん")
print(s.endswith("ん"))
```

```python
# df5の中身を辞書化
from collections import defaultdict
grouped = defaultdict(list)
for word in df5['見出し']:
    if word not in grouped[word[0]]:
        grouped[word[0]].append(word)
worddict_base = dict(grouped)
worddict_base
```

```python
# 単語数が少ないものをリスト化
wordlesses = []
for k, v in worddict_base.items():
    if len(v) < 5:
        print(k, v)
        wordlesses.append(k)
print(wordlesses)
```

```python
# 単語数が少ないものを補充...df4の段階で削ってるのでdf3で調整
df_less = df3.loc[df3['見出し'].str.startswith(tuple(wordlesses))]
# 「ん」で終わる単語を除外
df_less2 = df_less.loc[~df_less['見出し'].str.endswith('ん')]
# タイトル行を除外
df_less3 = df_less2['見出し'][1:]
df_less3
```

```python
# df_less3の中身をworddict_baseに補充
for word in df_less3:
    if word not in worddict_base[word[0]]:
        worddict_base[word[0]].append(word)
worddict_base
```

```python
# 単語数がどうなったか再確認...最低5単語は確保できたので10未満で確認している
wordlesses= [] #リストを初期化
for k, v in worddict_base.items():
    if len(v) < 10:
        print(k, v, len(v))
        wordlesses.append(k)
print(wordlesses)
```

```python
# どうも重複があるようなので調査
from collections import Counter

for k, v in worddict_base.items():
    counts = Counter(v)
    dups = [word for word, n in counts.items() if n > 1]
    if dups:
        print(dups)
```

```python
import json
with open('../data/worddict.json', mode="w", encoding="utf-8") as f: #withで開くとclose()を自動で行う
    json.dump(worddict_base, f, ensure_ascii=False, indent=2)
```
